## Statistical Testing

To analysis this data I used the Mann-Whitney U-Test test.

A two-tail p-value was also used as no prior assumptions are made about the contrast in the distributions of ridership on rainy and non-rainy days.

Null hypothesis
- NULL ( ENTRIESn_hourlyrainy ) and ( ENTRIESn_hourlynon âˆ’ rainy ) are same
- ALT ( ENTRIESn_hourlyrainy ) and ( ENTRIESn_hourlynon âˆ’ rainy ) differ by a location shift Î¼ and Î¼â‰ 0

p-critical value 
- Î± = 0.05

This test is applicable to the dataset because the Mann-Whitney U-Test tests the null hypothesis that the two samples being compared are derived from the same population. The Mannâ€“Whitney U test is a non-parametric test that is good for testing whether a particular population tends to have larger values than the other.

The Mann-Whitney U-Test tests the null hypothesis that the two samples being compared are derived from the same population. This null hypothesis allows us to test whether there is a statistically significant difference in ridership on rainy and non-rainy days (i.e., are the hourly entries derived from the same population). Furthermore, exploratory data analysis has shown that the data is not normally distributed. The Mann-Whitney U-Test does not assume normality of the data, making this test appropriate.


###Results

```python
import numpy as np

# means for each of the two samples under test.
rainy_day_mean = np.mean(turnstile_data[turnstile_data.rain == 1]['ENTRIESn_hourly'])
non_rainy_day_mean = np.mean(turnstile_data[turnstile_data.rain == 0]['ENTRIESn_hourly'])

print 'mean entries for rainny days are', np.round(rainy_day_mean, 3)
print 'mean entries for non-rainny days is', np.round(non_rainy_day_mean, 3)
```

```python
import scipy.stats

# p-value
man_u_value, p_value = scipy.stats.mannwhitneyu(turnstile_data[turnstile_data.rain == 1]['ENTRIESn_hourly'], 
                            turnstile_data[turnstile_data.rain == 0]['ENTRIESn_hourly'])

print 'Mann-Whitney U-Test:', man_u_value
p_value # One sided p value
print 'Two sided P-Value:', p_value * 2 # Two sided p value
```

###Significance

```python
#Significance Level 
alpha = 0.05

if (p_value * 2) < alpha:
    print 'Reject Null Hypothesis'
else:
    print 'Fail to Reject Null Hypothesis'
```

So with Î± = 0.05, and the p-value being so low we would Reject the Null meaning the distribution between rainy and non_rainy is statisticly different.

##Linear Regression

To compute the coefficients theta and produce prediction for ENTRIESn_hourly in your regression model, OLS was appropiate using the Statsmodels.

### Feature Selection

When I first started, I tried to include as many weather related variables as possible because I think people's choices of transportation highly depend on the condition of the weather.
But some of the variables are highly correlated, such as pressurei and meanpressurei, precipi, meanprecipi and rain etc. To avoid the problems can be caused by multicollinearity, I decided to include only one of each group of the highly correlated variables. 
To choose my features, I first needed to forget about the data and think what are things that would influence people to ride the subway, so I chose fog and conds as my dummy variables. Along with ENTRIESn, precipi, hour, tempi, UNIT, day_week, and pressurei in the model. I also used dmatrices from patsy, which does all the leg work for you, including adding constants.

```python
from patsy import dmatrices
from statsmodels import * 
import statsmodels.api as sm

y, X = dmatrices('ENTRIESn_hourly ~ precipi + C(hour) + tempi + \
                 UNIT + C(day_week) + pressurei + C(fog) + C(conds)', 
                 data=turnstile_data, return_type='dataframe')
model = sm.OLS(y, X)
results = model.fit()
print 'The R^2 of the model is', results.rsquared
```

This R^2 of 0.548 is not great with a skewness of 2.978. To analyze this low R^2 even further I plotted the residuals to see if their was structure to the residuals. 

```python
plt.figure.figsize = (10,6)
plt.scatter(turnstile_data['ENTRIESn_hourly'], results.resid, alpha=.05, c='r')
plt.axhline(y=0)
plt.title('Residual Scatter')
plt.xlabel('ENTRIESn_hourly')
plt.ylabel('Standardized Residuals')
plt.show()
```

This model was better but still wasn't satisfied with this R^2 and the structure of the residuals so I decided to take other means. After some tweeking and playing around with a Box-Cox transformation I discovered better, however still not great regression model.

```python
new_y = (y + 1)**0.25
updated_model = sm.OLS(new_y, X)
updated_results = updated_model.fit()
print 'The R^2 of the new updated model is', updated_results.rsquared
```

Once again I plot the residuals for this updated model

```python
plt.figure.figsize = (10,6)
plt.scatter(turnstile_data['ENTRIESn_hourly'], updated_results.resid, alpha=.05, c='r')
plt.axhline(y=0)
plt.title('Residual Scatter')
plt.xlabel('ENTRIESn_hourly')
plt.ylabel('Standardized Residuals')
plt.show()
```

This new model had an R^2 of 0.7454 and skewness of -0.278. Much better, but still not great.
Skewness is much better on this result. Skewness is a measurement of asymmetry of the distribution and of normal distribution the skewness is zero. So a good linear regression model should have the value of skewness close to zero.

#Future Predictions

```python
#Read again data
df = turnstile_data
#define a week in seconds
week = 7 * 24 * 60 * 60

df1 = df[['ENTRIESn_hourly', 'DATEn']].groupby('DATEn').sum()
df1.index.name = 'date'
df1.reset_index(inplace=True)
df1.columns = ['date', 'people_count']
df2 = df[['meanwspdi', 'meantempi', 'meanprecipi', 'meanpressurei', 'DATEn']].groupby('DATEn').mean()
df2.reset_index(inplace=True)

##join df1 and df2
joined = pd.concat([df1,df2], 1)
joined = joined.drop('DATEn', 1)
joined['date'] = pd.to_datetime(joined['date'])
joined['UnixTime'] = joined['date'].astype(numpy.int64) // 10**9
```

```python
#function to add fourier terms to data frame, n - number of terms
def add_term(df, n, col_time_name):
    for i in range(1,n+1):
        col_sin_name = "sin" + str(i)
        col_cos_name = "cos" + str(i)
        sin = lambda x: math.sin(2*math.pi*i*x/week)
        cos = lambda x: math.cos(2*math.pi*i*x/week)

        df[col_sin_name] = df[col_time_name].apply(sin)
        df[col_cos_name] = df[col_time_name].apply(cos)
    
    return df
```

```python
joined = add_term(joined, 2, 'UnixTime')
cols = list(joined.columns.values)
cols_x = [x for x in cols if x not in ['people_count', 'date', 'UnixTime']]
y = joined['people_count']
x = joined[cols_x]
x = sm.add_constant(x)
model = sm.OLS(y,x)
results = model.fit()
results.summary()
```

```python
from ggplot import *
pr = pd.Series(results.predict())
#Construct data with fitted values and real values
final1 = pd.DataFrame(dict(values = pr, date = joined['date']))
final2 = pd.DataFrame(dict(values = joined['people_count'], date = joined['date']))
final1['type'] = 'fitted values'
final2['type'] = 'real values'
final = pd.concat([final1, final2])

#visualization
plt.figure.figure = (10,6)
gg = ggplot(final, aes(x='date', y='values', color = 'type')) + geom_point() + geom_line() +\
ggtitle('Daily entries using Linear Regression') + ylim(0, 5000000)
plt.show()
print gg
```
