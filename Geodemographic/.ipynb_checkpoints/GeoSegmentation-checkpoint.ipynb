# Geodemographic Segmentation

#### Company Issue:

We are having a problem with customers leaving our bank. We need to find out why. Help us understand what we need to do.

Geodemographic segmentation is a multivariate statistical classification technique for discovering whether the individuals of a population fall into different groups by making quantitative comparisons of multiple characteristics with the assumption that the differences within any group should be less than the differences between groups.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
from scipy import stats
from __future__ import division
%matplotlib inline
style.use('ggplot')
```

Read the data in. It is a sample of Bank Customers with size of of 11000 rows and 14 columns

```python
dat = pd.read_csv('Churn_Data.csv')
dat.head()
```

Change the Gender Variable from `Male / Female` into a simple binary dummy variable: `1` for Male and `0` for Female

```python
gender_model = lambda row: [0,1] [row['Gender'] == 'Male']
dat['Gender'] = dat.apply(gender_model, axis = 1)
```

Create Dummy Variables for Geography

```python
geography_dummy = pd.get_dummies(dat.Geography)
dat = pd.concat([dat, geography_dummy], axis = 1)
```

Transform `Balance` to a more appropriate predictor, so I transform it to the Log10 of the Balance and simply add 1 to each Balance so the customers with 0 do not throw an error

```python
dat['LogBalance'] = np.log10(dat.Balance + 1)
```

Bin `Age`, since a continous variable is not appropriate here and customers actually have a higher churn rate in their mid-life rather than much later.  This allows the model to better understand the data

```python
bins = [17, 35, 42, 66, 75, 99]

dat['AgeBins'] = pd.cut(dat.Age, bins = bins)
age_dummies = pd.get_dummies(dat.AgeBins, prefix = 'Age')
dat = pd.concat([dat, age_dummies], axis = 1)

pd.DataFrame(dat.groupby('AgeBins')['Exited'].mean())
```

```python
dat.head()
```

Frequentist Approach to Chisquare AB Testing

```python
def AB_PLOT(exited1, exited2, stayed1, stayed2, category1, category2, on1, on2):
    
    data1 = [exited1, exited2]
    data2 = [stayed1, stayed2]

    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

    ## Absolute count
    ax1.bar(range(len(data1)), data1, label = on1, color = '#FF6600')
    ax1.bar(range(len(data2)), data2, bottom = data1, label = on2, color = 'b')
    plt.sca(ax1)
    plt.xticks([0.4, 1.4],  [category1, category2])
    ax1.set_ylabel("Count")
    plt.legend()

    ## Percent
    totals = [i + j for i,j in zip(data1, data2)]
    data1_rel = [i / j * 100 for  i,j in zip(data1, totals)]
    data2_rel = [i / j * 100 for  i,j in zip(data2, totals)]

    ax2.bar(range(len(data1_rel)), data1_rel, label = on1, color = '#FF6600')
    ax2.bar(range(len(data2_rel)), data2_rel, label = on2, bottom = data1_rel, color = 'b')
    plt.sca(ax2)
    plt.xticks([0.4, 1.4],  [category1, category2])
    plt.axhline(20, c = 'w')
    ax2.set_ylabel("Percentage %")
    plt.legend()

    plt.show()
    
def AB_Test(column, on1, on2):
    exited1 = dat[(dat[column] == 1) & (dat['Exited'] == 1)].shape[0]
    stayed1 = dat[(dat[column] == 1) & (dat['Exited'] == 0)].shape[0]

    exited2 = dat[(dat[column] == 0) & (dat['Exited'] == 1)].shape[0]
    stayed2 = dat[(dat[column] == 0) & (dat['Exited'] == 0)].shape[0]
    
    return AB_PLOT(exited1, exited2, stayed1, stayed2, on1, on2, 'Exited', 'Stayed')

# frequentist approach -- Numerical Output
def AB_Chisquare_Test(column):
    dat1 = dat[dat[column] == 0]
    dat2 = dat[dat[column] == 1]
    
    obs = [[len(dat1[dat1.Exited == 0]), len(dat1[dat1.Exited == 1])], 
       [len(dat2[dat2.Exited == 0]), len(dat2[dat2.Exited == 1])]]
    t_stat, p, dof, exp = stats.chi2_contingency(obs)
    return t_stat, p
```

```python
AB_Test('Gender', 'Male', 'Female')
t_stat, p = AB_Chisquare_Test('Gender')
print '-----Chi-Square Test Statistic-----: %.5f' % t_stat
print '-----P-Value-----: %.30f' % p
```

AB Test for IsActiveMember

```python
AB_Test('IsActiveMember', 'IsActive', 'NonActive')
t_stat, p = AB_Chisquare_Test('IsActiveMember')
print '-----Chi-Square Test Statistic-----: %.5f' % t_stat
print '-----P-Value-----: %.30f' % p
```

```python
AB_Test('HasCrCard', 'HasCrCard', 'NoCrCard')
t_stat, p = AB_Chisquare_Test('HasCrCard')
print '-----Chi-Square Test Statistic-----: %.5f' % t_stat
print '-----P-Value-----: %.30f' % p
```

### Bayesian A/B TEST

```python
import pymc as pm

def bayes_AB(column):
    # get count and len for both priors
    data_A = dat[(dat[column] == 0) & (dat.Exited == 1)].shape[0] 
    data_A_len = dat[dat[column] == 0].shape[0]
    data_B = dat[(dat[column] == 1) & (dat.Exited == 1)].shape[0]
    data_B_len = dat[dat[column] == 1].shape[0]
    
    # set up numpy array of ones and zeros 
    data_A = np.r_[[0] * (data_A_len - data_A), [1] * data_A]
    data_B = np.r_[[0] * (data_B_len - data_B), [1] * data_B]
    
    # create priors, have no assumption so use uniform priors
    p_A = pm.Uniform('p_A', lower = 0, upper = .99)
    p_B = pm.Uniform('p_B', lower = 0, upper = .99)
    
    # set up determistic variable to calc delta(difference in samples)
    @pm.deterministic
    def delta(p_A=p_A, p_B=p_B):
        return p_B - p_A
    
    # create likelihoods with Bernoulli dist
    obs_A = pm.Bernoulli('obs_A', p_A, value = data_A, observed = True)
    obs_B = pm.Bernoulli('obs_B', p_B, value = data_B, observed = True)
    
    return pm.MCMC([p_A, p_B, delta, obs_A, obs_B])
```

```python
def bayes_plot_dist(col1, col2, delta, col1_name, col2_name):

    plt.figure(figsize=(10, 10))

    plt.subplot(3,1,1)
    plt.hist(col1, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of ${}$".format(col1_name), color="#A60628", normed=True)
    plt.title("Posterior distributions of ${}$".format(col1_name))
    plt.legend()

    plt.subplot(3,1,2)
    plt.hist(col2, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of ${}$".format(col2_name), color="#467821", normed=True)
    plt.title("Posterior distributions of ${}$".format(col2_name))
    plt.legend()

    plt.subplot(3,1,3)
    plt.hist(delta, histtype='stepfilled', bins=25, alpha=0.85,
             label="posterior of $delta$", color="#7A68A6", normed=True)
    plt.title("Posterior distributions of $delta$ unknowns")
    plt.axvline(0, linestyle = '--')
    plt.legend()

    plt.tight_layout()
    
    print 'Probability of {} exiting LESS than {} exiting: {}'.format(col1_name,col2_name,(delta_samples < 0).mean())
    print 'Probability {} exiting MORE than {} exiting: {}'.format(col2_name,col1_name,(delta_samples > 0).mean())
```

#### Gender

## Modeling

Set up features to include in the model. Here for the categorical variables, `Gender` is fine. For `Age`, I use the last age bracket of 75 - 99 as the reference level since it has the lowest chance of churn. And as for `Geography` I only include Germany in the final model as originally I used both Germany and Spain however Spain was not significant as 0.05. I set up the final model with X and y with X as the predictors and y and the dependant, and add a constant to the features.

```python
import statsmodels.api as sm

X = dat[['CreditScore', 'Gender', 'Age_(17, 35]', 'Age_(35, 42]', 'Age_(42, 66]', 'Age_(66, 75]',
         'Tenure', 'NumOfProducts', 'IsActiveMember', 'Germany', 'LogBalance']]
y = dat.Exited
```

Correlation Plot

```python
plt.figure(figsize=(7,7))
sns.corrplot(X)
plt.tight_layout()
```

I fit a Logistic Regression model to the training data, I used a simple Logistic Regression model and not a more complicated technique because in this situation it is more important to have a more interpretable result rather than a more predictive model using an ensemble method or SVC. 

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1000, random_state = 100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import chi2

clf = LogisticRegression(fit_intercept=True)
clf.fit(X_train, y_train)

scores, pvalues = chi2(X_train, y_train)

coef = pd.DataFrame(zip(clf.coef_.T, np.exp(clf.coef_).T))
coef.columns = ['Coef_','Odds_Ratio_']
coef['P_values'] = pvalues
coef.index = X_train.columns

sns.set()
coef.Coef_ = coef.Coef_.astype(np.float64)
coef.Odds_Ratio_ = coef.Odds_Ratio_.astype(np.float64)

columns = ['Odds_Ratio_']

sns.heatmap(coef, annot=True, fmt = '.4g')
plt.show()
```

Focus your eyes on the `Odds Ratio` column, the middle column in the dataframe as this is where all the info is. I have color coded the the importance of each predictors.  The higher the `Odds Ratio` the more likely a person will leave the bank. So based on the output the highest churn rate will occur when a person is between the ages of 42 and 65, with the second highest being customers who are in the `Germany` branch.  So as a bank, you need to address these people first with the most being people from both groups.

#### Confussion Matrix Train and Test Set

```python
from sklearn.metrics import confusion_matrix

pred_train = clf.predict(X_train)
pred_test = clf.predict(X_test)
print '---- Baseline for train and test set: {}, {} ----'.format(np.mean(y_train == 0), np.mean(y_test == 0))
plt.figure(figsize=(7,3))
ax = plt.subplot(1,2,1)
sns.heatmap(confusion_matrix(y_train, pred_train), annot = True, fmt = '')
plt.title('Training Set Accuracy: {}'.format(clf.score(X_train, y_train)))
ax = plt.subplot(1,2,2)
sns.heatmap(confusion_matrix(y_test, pred_test), annot = True, fmt = '')
plt.title('Testing Set Accuracy: {}'.format(clf.score(X_test, y_test)))
plt.tight_layout()
```

#### ROC Curves for Train and Test Sets

True positive rate (Sensitivity) is plotted in function of the false positive rate (100-Specificity) for different cut-off points of a parameter.

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fig, ax = plt.subplots(1,2)

x = np.linspace(0, 1, 10)
y = x

pred_prob = clf.predict_proba(X_train)[:,1]
pred_test_prob = clf.predict_proba(X_test)[:,1]
# train ROC
fp, tp, _ = roc_curve(y_train, pred_prob)
ax[0].plot(fp, tp)
ax[0].plot(x, y, '--')
ax[0].set_title('Train AUC: {}'.format(roc_auc_score(y_train, pred_train)))
# test ROC
fp, tp, _ = roc_curve(y_test, pred_test_prob)
ax[1].plot(fp, tp)
ax[1].plot(x,y, '--')
ax[1].set_title('Test AUC: {}'.format(roc_auc_score(y_test, pred_test)))
plt.tight_layout()
```

#### CAP Curve

```python
forcast = pd.DataFrame(zip(y_train.index, y_train, clf.predict_proba(X_train)[:,1]))
forcast.columns = ['RowNumber', 'Exited', 'Pred_Prob']
forcast.sort('Pred_Prob', inplace=True, ascending=False)
forcast.to_csv('forcast_good.csv', index = False)
```

The Cumulative Accuracy Profiles (CAP) curve is an approach that allows a business to target their audience. This is built in excel as their is an easy template to create this image and would be a pain to do it with code. This image is read as: the red line indicates random guessing and the yellow is what the classifer predicts.  If the bank went after the customers that the logistic regression predicted as the highest probabilities the bank would be able to target much better than the base line. You can compare the two lines and see how much influence the classifer has on the data.
